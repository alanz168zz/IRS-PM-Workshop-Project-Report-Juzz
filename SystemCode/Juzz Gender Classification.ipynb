```python
# EDA package
import pandas as pd
import numpy as np
```

```python
# ML packages
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction import DictVectorizer
#from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
# Load the dataset
df = pd.read_csv('drive/My Drive/JuzzChatBot/100names.csv')
df.head()
```

```python
df.size
```

```python
#Data Cleaning
# Check for column name consistency
df.columns
```

```python
#Data types
df.dtypes
```

```python
# Check for missing values
df.isnull().isnull().sum()
```

```python
# Number of female names
df[df.sex == 'female'].size
```

```python
# Number of Male names
df[df.sex == 'male'].size
```

```python
df_name = df
```

```python
#Replacing all female and male with 0 and 1 
df_name.sex.replace({'female':0,'male':1}, inplace=True)
```

```python
df_name.sex.unique()
```

```python
df_name.dtypes
```

```python
Xfeatures = df_name['names']
```

```python
# Feature Extraction
cv = CountVectorizer()
X = cv.fit_transform(Xfeatures)
```

```python
cv.get_feature_names()
```

```python
from sklearn.model_selection import train_test_split
```

```python
#Features
X
#Labels
y = df_name.sex
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.33,random_state = 42)
```

```python
# Naive Baynes Classifier
from sklearn.naive_bayes import MultinomialNB
clf = MultinomialNB()
clf.fit(X_train,y_train)
clf.score(X_test,y_test)
```

```python
# Accuracy of our model
print("Accuracy of Model",clf.score(X_test,y_test)*100,"%")
```

```python
#Accuracy of our model
print("Accuracy of model", clf.score(X_train,y_train)*100,"%")
```

```python
#Sample1 Prediction
sample_name = ["Mark"]
vect =cv.transform(sample_name).toarray()
```

```python
#Female is 0, Male is 1
clf.predict(vect)
```
